```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img1 = cv2.imread('./img/Perro1.jpg')
img2 = cv2.imread('./img/confidencial.png')
img1.shape
img2.shape
```

```python
# plt.imshow(img1)
img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2BGR)
# img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2BGR)
plt.imshow(img1)
# plt.imshow(img2)
```

```python
# plt.imshow(img1)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
# plt.imshow(img1)
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.3,gamma=0)

plt.imshow(blended)
```

```python
img1_2 = cv2.imread('./img/Perro1.jpg')
img2_2 = cv2.imread('./img/confidencial.png')
img2_2 =cv2.resize(img2_2,(600,600))
img1_2 = cv2.cvtColor(img1_2, cv2.COLOR_BGR2RGB)
img2_2 = cv2.cvtColor(img2_2, cv2.COLOR_BGR2RGB)
```

```python
large_img = img1_2
small_img = img2_2
```
